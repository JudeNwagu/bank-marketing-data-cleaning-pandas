The numerical features reveal several critical data quality and analytical considerations. The pdays variable is dominated by the placeholder value 999, appearing in 96.3% of rows, which renders its statistical measures meaningless; this value must be replaced with NaN before any analysis. Only 3.7% of customers were previously contacted, forming a small but highly valuable returning-customer segment. The duration column should be dropped entirely, as it represents post-call information not available at prediction time, includes invalid cases (duration = 0), and contains extreme outliers (up to 4,918 seconds) that would heavily skew analysis. In contrast, age is clean and well-distributed, ranging from 17 to 98 with a mean of 40, making it suitable for segmentation without additional cleaning.

```python
before = len(df)
df = df.drop_duplicates().reset_index(drop=True)
after = len(df)
print(f"Rows removed: {before - after}")
print(f"Rows remaining: {after:,}")
```

```python
df.drop(columns=['duration'], inplace=True)
print("duration column dropped.")
print("Remaining columns:", df.shape[1])
```

```python
# Only replace "unknown" in the columns where it is a missing placeholder. Do NOT touch poutcome — "nonexistent" there is valid documented dat

unknown_cols = ['job','marital','education',
                'default','housing','loan']

for col in unknown_cols:
    count = (df[col] == 'unknown').sum()
    df[col] = df[col].replace('unknown', np.nan)
    print(f"{col}: {count} unknowns → NaN")

print()
print("NaN counts after replacement:")
print(df[unknown_cols].isnull().sum())
```

```python
# Under 5% missing → fill with mode (most frequent). Over 5% missing → label as "Unknown" so Power BI can show it transparently.

threshold = 0.05

for col in unknown_cols:
    pct = df[col].isnull().mean()
    if pct < threshold:
        fill_val = df[col].mode()[0]
        df[col] = df[col].fillna(fill_val)      # fixed
        print(f"{col}: {pct*100:.1f}% missing → filled with '{fill_val}'")
    else:
        df[col] = df[col].fillna('Unknown')     # fixed
        print(f"{col}: {pct*100:.1f}% missing → kept as 'Unknown'")
```

```python
df['pdays'] = df['pdays'].replace(999, np.nan)

df['previously_contacted'] = df['pdays'].notna().map(
    {True: 'Yes', False: 'No'}
)

print("pdays NaN count:", df['pdays'].isnull().sum())
print(df['previously_contacted'].value_counts())
```

```python
#### Standardise text values "Admin." and "admin." as different categories and often cause confusion on our code or when using it on powerbi best is Strip and lowercase everything.
```

```python
# Identify all columns with 'object' (string) data types
str_cols = df.select_dtypes(include=['object']).columns

for col in str_cols:
    df[col] = df[col].str.strip().str.lower()

# Remove trailing dot from "admin." specifically
df['job'] = df['job'].str.replace(r'\.$', '', regex=True)

print(df['job'].unique())
```

```python
# Officially documented as a directive: "divorced" includes widowed. Rename for accurate Power BI reporting.
df['marital'] = df['marital'].replace(
    'divorced', 'divorced/widowed'
)
print(df['marital'].value_counts())
```

```python
# Replace dot notation('basic.4y') with readable labels for Power BI axis and tooltips.
edu_labels = {
    'basic.4y':            'Basic (4 years)',
    'basic.6y':            'Basic (6 years)',
    'basic.9y':            'Basic (9 years)',
    'high.school':         'High School',
    'professional.course': 'Professional Course',
    'university.degree':   'University Degree',
    'illiterate':          'Illiterate',
    'unknown':             'Unknown'
}
df['education'] = df['education'].replace(edu_labels)
print(df['education'].value_counts())
```

```python
# Education is ordinal — assign a numeric rank so Power BI sorts it correctly on charts instead of alphabetically.
edu_order = {
    'illiterate':          0,
    'Basic (4 years)':     1,
    'Basic (6 years)':     2,
    'Basic (9 years)':     3,
    'High School':         4,
    'Professional Course': 5,
    'University Degree':   6,
    'Unknown':             7
}
df['education_rank'] = df['education'].map(edu_order)
print(df[['education','education_rank']].drop_duplicates()
        .sort_values('education_rank'))
```

```python
# Age as raw integers makes poor Power BI charts. Grouped bands make clean bar charts and slicers.

df['age_band'] = pd.cut(
    df['age'],
    bins=[0, 25, 35, 45, 55, 65, 100],
    labels=['Under 25','25–34','35–44',
            '45–54','55–64','65+']
)
print(df['age_band'].value_counts().sort_index())
```

```python
# Without sorting columns, Power BI will sort months alphabetically (apr, aug, dec…) instead of chronologically (jan, feb, mar…).
month_order = {'jan':1,'feb':2,'mar':3,'apr':4,
               'may':5,'jun':6,'jul':7,'aug':8,
               'sep':9,'oct':10,'nov':11,'dec':12}

day_order = {'mon':1,'tue':2,'wed':3,
             'thu':4,'fri':5}

df['month_order']   = df['month'].map(month_order)
df['day_order']     = df['day_of_week'].map(day_order)

print(df[['month','month_order']].drop_duplicates()
        .sort_values('month_order'))
print()
print(df[['day_of_week','day_order']].drop_duplicates()
        .sort_values('day_order'))
```

```python
# Keep the text column for Power BI labels. Add a numeric 0/1 column for KPI cards, conversion rate calculations, and any future modelling.

df['subscribed'] = df['y'].map({'yes': 1, 'no': 0})

print(df['subscribed'].value_counts())
print(f"\nConversion rate: {df['subscribed'].mean()*100:.1f}%")
```

```python
# Dots in column names break DAX formulas in Power BI silently. Rename all four economic columns now.
df.rename(columns={
    'emp.var.rate':   'emp_var_rate',
    'cons.price.idx': 'cons_price_idx',
    'cons.conf.idx':  'cons_conf_idx',
    'nr.employed':    'nr_employed'
}, inplace=True)

print("Renamed columns:")
print([c for c in df.columns if '_' in c])
```
